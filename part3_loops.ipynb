## 📝 Notes

### `for` Loop
Iterates over a sequence (list, string, range, etc.):
```python
for i in range(5):
    print(i)  # 0 1 2 3 4
```

### `while` Loop
Runs as long as a condition is `True`:
```python
i = 0
while i < 5:
    print(i)
    i += 1
```
⚠️ Always make sure the condition eventually becomes `False`, or you'll get an **infinite loop**!

### `range()`
| Syntax | Meaning |
|--------|---------|
| `range(5)` | `0, 1, 2, 3, 4` |
| `range(2, 6)` | `2, 3, 4, 5` |
| `range(0, 10, 2)` | `0, 2, 4, 6, 8` |
| `range(10, 0, -1)` | `10, 9, 8, ... 1` |

⚠️ `range()` is **exclusive** of the stop value — `range(1, 6)` gives `1` to `5`, not `6`.

### `enumerate()`
Returns `(index, value)` pairs — avoids manual index tracking:
```python
for i, val in enumerate(["a", "b", "c"]):
    print(i, val)  # 0 a, 1 b, 2 c

# Start index from 1:
for i, val in enumerate(["a", "b", "c"], start=1):
    print(i, val)  # 1 a, 2 b, 3 c
```

### `break` vs `continue`
| Keyword | Effect |
|---------|--------|
| `break` | **Exits** the loop entirely |
| `continue` | **Skips** the rest of the current iteration and moves to the next |

```python
# break example:
for i in range(10):
    if i == 5:
        break       # stops at 5, never prints it
    print(i)        # 0 1 2 3 4

# continue example:
for i in range(5):
    if i == 3:
        continue    # skips 3, keeps going
    print(i)        # 0 1 2 4
```

### Accumulator Pattern
Always **initialise** your accumulator variable before the loop:
```python
total = 0          # ← initialise BEFORE the loop
for i in range(1, 6):
    total += i
print(total)       # 15
```
⚠️ Forgetting to initialise causes a `NameError`!

### `while True` + `break`
A clean pattern for loops that run until a condition is met:
```python
i = 0
while True:
    i += 1
    print(i)
    if i == 7:
        break
```

## 🔍 Part A: Output Questions

### Q1 — `range(start, stop, step)`
**Question:** What does this output?

**Answer:** `2 4 6 8` — `range(2, 10, 2)` starts at `2`, stops before `10`, steps by `2`. The stop value is always **excluded**.

````python
for i in range(2, 10, 2):
    print(i)
# Output: 2 4 6 8  (10 is excluded — range stop is always exclusive)
````

### Q2 — `enumerate()`
**Question:** What does this output?

**Answer:** `0 a`, `1 b`, `2 c` — `enumerate()` returns `(index, value)` pairs starting from `0` by default.

````python
for i, val in enumerate(["a", "b", "c"]):
    print(i, val)
# Output:
# 0 a
# 1 b
# 2 c
````

### Q3 — `continue` in a `while` loop
**Question:** What does this output?

**Answer:** `1 2 4 5` — `continue` skips the `print` for that iteration but the loop keeps going. It does **not** stop the loop.

````python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue    # skips print(3), loop continues
    print(i)
# Output: 1 2 4 5

# ⚠️ Common mistake: confusing continue with break
# continue → skip this iteration
# break    → exit the loop entirely
````

### Q4 — `break` in a `for` loop
**Question:** What does this output?

**Answer:** `0 1 2 3 4` — `break` exits the loop immediately when `i == 5`, so `5` is never printed.

````python
for i in range(10):
    if i == 5:
        break       # exits loop immediately — 5 is never printed
    print(i)
# Output: 0 1 2 3 4
````

### Q5 — Accumulator pattern
**Question:** What does this output?

**Answer:** `15` — `range(1, 6)` gives `1, 2, 3, 4, 5` and their sum is `15`.

````python
total = 0
for i in range(1, 6):
    total += i
print(total)  # Output: 15  (1+2+3+4+5)
````

## 💻 Part B: Coding Exercises

### Q6 — `while` loop with `break`
**Task:** Print numbers from `1` upward and stop when you hit a number divisible by `7`.

**Key concept:** `while True` + `break` is a clean pattern when you don't know upfront how many iterations you need.

````python
i = 0
while True:
    i += 1
    print(i)
    if i % 7 == 0:
        break
# Output: 1 2 3 4 5 6 7
````

### Q7 — Times table with `range()`
**Task:** Print the times table of 6 from 6×1 to 6×10.

**Key concept:** `range(1, 11)` gives `1` through `10` — multiply each by `6`.

````python
for i in range(1, 11):
    print(i * 6)
# Output: 6 12 18 24 30 36 42 48 54 60

# Bonus — formatted output:
for i in range(1, 11):
    print(f"6 x {i} = {i * 6}")
````

### Q8 — `enumerate()` with custom start
**Task:** Print each fruit with its position number starting from `1`.

**Key concept:** Use `enumerate(list, start=1)` and f-strings for clean formatted output.

````python
fruits = ["apple", "banana", "cherry"]

for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
# Output:
# 1. apple
# 2. banana
# 3. cherry

# ⚠️ Common mistake — print(i, ".", fruit) adds unwanted spaces:
# 1 . apple  ← wrong
# Always use f-strings for formatted output!
````

### Q9 — FizzBuzz
**Task:** For numbers `1` to `20`, print `"FizzBuzz"` if divisible by both 3 and 5, `"Fizz"` if by 3, `"Buzz"` if by 5, else print the number.

**Key concept:** Check the combined case (`% 15`) first — otherwise it gets swallowed by `Fizz` or `Buzz`.

````python
for i in range(1, 21):
    if i % 15 == 0:        # check combined case FIRST
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

### Q10 — Sum of even numbers
**Task:** Use a `for` loop to print the sum of all even numbers from `1` to `50`.

**Key concept:** Always initialise accumulator variables **before** the loop, or you'll get a `NameError`.

````python
total = 0              # ✅ always initialise BEFORE the loop

for i in range(1, 51):
    if i % 2 == 0:
        total += i

print(total)           # Output: 650

# Bonus — cleaner with range step:
total = 0
for i in range(2, 51, 2):  # step by 2 to hit only even numbers
    total += i
print(total)               # Output: 650
````

## 🗂️ Summary & Key Takeaways

| Concept | Key Point |
|--------|-----------|
| `range(start, stop, step)` | Stop is always **exclusive** — `range(1, 6)` gives `1` to `5` |
| `enumerate()` | Returns `(index, value)` pairs — use `start=1` to begin from 1 |
| `break` | **Exits** the loop entirely — the triggering value is never processed |
| `continue` | **Skips** the rest of the current iteration — loop keeps going |
| `while True` + `break` | Clean pattern when you don't know how many iterations you need |
| Accumulator pattern | Always initialise (`total = 0`) **before** the loop |
| f-strings in loops | Use `f"{i}. {val}"` instead of `print(i, ".", val)` to avoid unwanted spaces |
| FizzBuzz rule | Always check the combined condition (`% 15`) **first** |
| Even numbers shortcut | `range(2, n+1, 2)` steps through even numbers directly — no `if` needed |

### ⚠️ Common Mistakes to Avoid
- Confusing `break` and `continue` — `break` stops everything, `continue` just skips one iteration
- Forgetting to initialise accumulator variables before the loop
- Using `print(a, ".", b)` instead of f-strings causing unwanted spaces
- Wrong `range()` bounds — remember stop is **always exclusive**
- Wrong indentation in `if/elif/else` inside loops — all branches must align
