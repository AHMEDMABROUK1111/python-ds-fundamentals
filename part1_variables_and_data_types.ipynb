## 📝 Notes

### Data Types
| Type | Example | Description |
|------|---------|-------------|
| `int` | `x = 10` | Whole numbers |
| `float` | `x = 3.14` | Decimal numbers |
| `str` | `x = "hello"` | Text, always in quotes |
| `bool` | `x = True` | `True` or `False` only |

### Key Rules
- Use `type(x)` to check the type of any variable
- `/` always returns a `float` in Python 3 — use `//` for integer division
- You **cannot** concatenate `str` and `int` with `+` — convert with `str()` or use f-strings
- Mixing `int` and `float` in operations returns a `float` (upcasting)
- `bool` is a subclass of `int`: `True == 1` and `False == 0`

### Falsy Values
These all evaluate to `False` when passed to `bool()`:
- `0`, `0.0`, `""`, `None`, `False`

### Valid Variable Names
- ✅ Must start with a letter or `_`
- ✅ Can contain letters, numbers, underscores
- ❌ Cannot start with a number (`2name`)
- ❌ Cannot use hyphens (`my-var`)
- ❌ Cannot use reserved keywords (`class`, `if`, `for`, etc.)

### Type Conversion
```python
int("42")     # → 42
float("3.14") # → 3.14
str(10)       # → "10"
bool(0)       # → False
```

## ✏️ Questions & Answers

### Q1 — Division operator & float output
**Question:** What will the following code output?

**Answer:** `3.3333333333333335`

In Python 3, `/` always returns a `float`. Use `//` for integer (floor) division.

````python
x = 10
y = 3
print(x / y)   # Output: 3.3333333333333335  → true division, always float
print(x // y)  # Output: 3                   → floor division, returns int
````

### Q2 — String vs Integer
**Question:** What is the type of `x = "42"`?

**Answer:** `str`

Quotes always make a value a string, even if it looks like a number. Use `int(x)` to convert it to an integer.

````python
x = "42"
print(type(x))      # Output: <class 'str'>

y = int(x)
print(type(y))      # Output: <class 'int'>
````

### Q3 — `type()` with booleans
**Question:** What does `print(type(True))` output?

**Answer:** `<class 'bool'>`

`bool` is a subclass of `int` in Python, so `True == 1` and `False == 0`.

````python
x = True
print(type(x))    # Output: <class 'bool'>

# bool is a subclass of int
print(True == 1)  # Output: True
print(False == 0) # Output: True
````

### Q4 — String + Integer concatenation
**Question:** What happens when you do `"My name is " + 25`?

**Answer:** `TypeError`

You cannot concatenate `str` and `int` with `+`. Use `str()` to convert, or use an f-string.

````python
name = "Alice"
age = 25

# ❌ This raises TypeError — cannot concatenate str and int:
# print("My name is " + age)

# ✅ Fix 1 — convert int to str:
print("My name is " + str(age))  # Output: My name is 25

# ✅ Fix 2 — f-string (recommended):
print(f"My name is {age}")       # Output: My name is 25
````

### Q5 — Type upcasting (int + float)
**Question:** What is the type of `5 + 2.0`?

**Answer:** `<class 'float'>`

When mixing `int` and `float`, Python automatically upcasts the result to `float`.

````python
x = 5
y = 2.0
print(x + y)        # Output: 7.0
print(type(x + y))  # Output: <class 'float'>
````

### Q6 — Valid variable names
**Question:** Which is a valid Python variable name: `2name`, `my-variable`, `_my_var`, or `class`?

**Answer:** `_my_var`

Variable names must start with a letter or `_`, can't use `-`, and can't be reserved keywords.

````python
# ✅ Valid variable names:
_my_var   = 10
myVar     = 20
my_var_2  = 30
print(_my_var, myVar, my_var_2)  # Output: 10 20 30

# ❌ Invalid — would raise SyntaxError:
# 2name = 10        → cannot start with a number
# my-variable = 10  → hyphen is interpreted as minus operator
# class = 10        → 'class' is a reserved keyword
````

### Q7 — `bool()` conversion & falsy values
**Question:** What does `bool(0)` return?

**Answer:** `False`

`0`, `0.0`, and `""` are all falsy in Python. Any non-zero number or non-empty string is truthy.

````python
# Falsy values:
print(bool(0))    # False
print(bool(0.0))  # False
print(bool(""))   # False

# Truthy values:
print(bool(1))    # True
print(bool(-5))   # True
print(bool("hi")) # True
````

### Q8 — `int()` type conversion
**Question:** What does `10 * int("5")` output?

**Answer:** `50`

`int("5")` converts the string `"5"` to the integer `5`, then `10 * 5 = 50`.

````python
x = 10
y = "5"
print(x * int(y))  # Output: 50
````
