## 📦 Part 1: Variables & Data Types

### Q1 — Division operator & float output
**Question:** What will the following code output?

**Answer:** `3.3333333333333335` — In Python 3, `/` always returns a `float`. Use `//` for integer (floor) division.

```python
x = 10
y = 3
print(x / y)  # Output: 3.3333333333333335
```

### Q2 — String vs Integer
**Question:** What is the type of `x = "42"`?

**Answer:** `str` — Quotes always make a value a string, even if it looks like a number. Use `int(x)` to convert.

```python
x = "42"
print(type(x))  # Output: <class 'str'>
```

### Q3 — `type()` with booleans
**Question:** What does `print(type(True))` output?

**Answer:** `<class 'bool'>` — Note: `bool` is a subclass of `int` in Python, so `True == 1` and `False == 0`.

```python
x = True
print(type(x))  # Output: <class 'bool'>
print(True == 1)  # Output: True (bool is a subclass of int)
```

### Q4 — String + Integer concatenation
**Question:** What happens when you do `"My name is " + 25`?

**Answer:** `TypeError` — You cannot concatenate `str` and `int` with `+`. Convert using `str()` or use an f-string.

```python
name = "Alice"
age = 25

# ❌ This raises TypeError:
# print("My name is " + age)

# ✅ Fix 1 — str() conversion:
print("My name is " + str(age))

# ✅ Fix 2 — f-string (preferred):
print(f"My name is {age}")
```

### Q5 — Type upcasting (int + float)
**Question:** What is the type of `5 + 2.0`?

**Answer:** `<class 'float'>` — When mixing `int` and `float`, Python automatically upcasts the result to `float`.

```python
x = 5
y = 2.0
print(type(x + y))  # Output: <class 'float'>
print(x + y)        # Output: 7.0
```

### Q6 — Valid variable names
**Question:** Which is a valid Python variable name: `2name`, `my-variable`, `_my_var`, or `class`?

**Answer:** `_my_var` — Variable names must start with a letter or `_`, can't use `-` (it's an operator), and can't use reserved keywords like `class`.

```python
# ✅ Valid variable names:
_my_var = 10
myVar = 20
my_var_2 = 30

# ❌ Invalid (would raise SyntaxError):
# 2name = 10       → starts with a number
# my-variable = 10 → hyphen is a minus operator
# class = 10       → reserved keyword
```

### Q7 — `bool()` conversion & falsy values
**Question:** What does `bool(0)` return?

**Answer:** `False` — In Python, `0`, `0.0`, and `""` are all falsy. Any non-zero number or non-empty string is truthy.

```python
print(bool(0))    # False
print(bool(0.0))  # False
print(bool(""))   # False
print(bool(1))    # True
print(bool(-5))   # True
print(bool("hi")) # True
```

### Q8 — `int()` type conversion
**Question:** What does `10 * int("5")` output?

**Answer:** `50` — `int("5")` converts the string `"5"` to the integer `5`, then `10 * 5 = 50`.

```python
x = 10
y = "5"
print(x * int(y))  # Output: 50
```

## 🔀 Part 2: Control Flow — `if` / `elif` / `else`

### Q1 — Positive, Negative, or Zero
Print `"positive"`, `"negative"`, or `"zero"` depending on `x`.

```python
x = -5

if x < 0:
    print("negative")
elif x > 0:
    print("positive")
else:
    print("zero")
```

### Q2 — Teenager Check
Print `"teenager"` if age is between 13 and 17 (inclusive), otherwise `"not a teenager"`.

💡 **Tip:** Python supports chained comparisons: `13 <= age <= 17`

```python
age = 15

if 13 <= age <= 17:  # Chained comparison — Pythonic!
    print("teenager")
else:
    print("not a teenager")
```

### Q3 — Login System
Print `"access granted"` if username is `"admin"` AND password is `"1234"`, otherwise `"access denied"`.

💡 **Key rule:** Use `==` to compare, `=` is for assignment only.

```python
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("access granted")
else:
    print("access denied")
```

### Q4 — Score Validator
Print `"invalid score"` if below 0 or above 100, `"pass"` if 50+, otherwise `"fail"`.

💡 **Key lesson:** Always check edge cases FIRST — Python stops at the first `True` condition.

```python
score = 85

if score < 0 or score > 100:  # ✅ Edge case checked FIRST
    print("invalid score")
elif score >= 50:
    print("pass")
else:
    print("fail")
```

### Q5 — FizzBuzz
Print `"fizzbuzz"` if divisible by both 3 and 5, `"fizz"` if by 3, `"buzz"` if by 5, else print the number.

💡 **Key lesson:** `%` is the modulo (remainder) operator. `n % 3 == 0` means n is divisible by 3.

```python
n = 15

if n % 15 == 0:   # Divisible by both 3 and 5 — checked FIRST
    print("fizzbuzz")
elif n % 3 == 0:
    print("fizz")
elif n % 5 == 0:
    print("buzz")
else:
    print(n)
```

### Q6 — Letter in Word
Print `"yes"` if the word contains the letter `"a"`, otherwise `"no"`.

```python
word = "banana"

if "a" in word:  # 'in' checks membership in strings and lists
    print("yes")
else:
    print("no")
```

### Q7 — Both Even, Both Odd, or Mixed
Print `"both even"`, `"both odd"`, or `"mixed"` based on two numbers.

```python
a = 4
b = 7

if a % 2 == 0 and b % 2 == 0:
    print("both even")
elif a % 2 != 0 and b % 2 != 0:
    print("both odd")
else:
    print("mixed")
```

### Q8 — Not Negative
Print `"yes"` if the number is not negative, otherwise `"no"`.

```python
x = 7

if x >= 0:
    print("yes")
else:
    print("no")

# Alternative using 'not':
# if not x < 0:
```

### Q9 — Driving Eligibility
Print `"eligible"` if age is 18+ AND has a valid licence, otherwise `"not eligible"`.

💡 **Tip:** Since `has_licence` is already a boolean, you don't need `== True` — just use it directly.

```python
age = 20
has_licence = True

if age >= 18 and has_licence:  # No need for '== True'
    print("eligible")
else:
    print("not eligible")
```

### Q10 — Grade Calculator
Print `"grade A"` for 90+, `"grade B"` for 80–89, `"grade C"` for 70–79, `"fail"` for below 70.

```python
score = 85

if score >= 90:
    print("grade A")
elif 80 <= score <= 89:  # Chained comparison
    print("grade B")
elif 70 <= score <= 79:
    print("grade C")
else:
    print("fail")
```

## 📝 Key Takeaways

| Concept | Notes |
|--------|-------|
| `/` vs `//` | `/` always returns float; `//` is integer division |
| `=` vs `==` | `=` assigns, `==` compares |
| `%` operator | Returns remainder — use for divisibility checks |
| `str + int` | TypeError — convert with `str()` or use f-strings |
| Falsy values | `0`, `0.0`, `""`, `None`, `False` all evaluate to `False` |
| Edge cases | Always check edge cases first in `if/elif` chains |
| Boolean check | `if has_licence:` is cleaner than `if has_licence == True:` |
| Chained comparisons | `13 <= age <= 17` is valid and Pythonic |
| `in` operator | Checks membership in strings, lists, etc. |
| Variable names | Must start with a letter or `_`, no hyphens, no keywords |
