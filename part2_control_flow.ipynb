## 📝 Notes

### `if` / `elif` / `else` Structure
```python
if condition:
    # runs if condition is True
elif another_condition:
    # runs if first was False and this is True
else:
    # runs if all above were False
```
⚠️ **Indentation is mandatory in Python** — always indent the body with 4 spaces.

### Comparison Operators
| Operator | Meaning | Example |
|----------|---------|--------|
| `==` | Equal to | `x == 5` |
| `!=` | Not equal to | `x != 5` |
| `>` | Greater than | `x > 5` |
| `<` | Less than | `x < 5` |
| `>=` | Greater than or equal | `x >= 5` |
| `<=` | Less than or equal | `x <= 5` |

### Logical Operators
| Operator | Meaning | Example |
|----------|---------|--------|
| `and` | Both must be True | `x > 0 and x < 10` |
| `or` | At least one must be True | `x < 0 or x > 100` |
| `not` | Inverts the condition | `not x < 0` |

### Key Tips
- `=` assigns, `==` compares — easy to confuse!
- `%` is the modulo operator — returns the remainder e.g. `10 % 3 == 1`
- Use `%` to check divisibility: `n % 3 == 0` means n is divisible by 3
- Always check **edge cases first** — Python stops at the first `True` condition
- Use chained comparisons for ranges: `13 <= age <= 17` instead of `age >= 13 and age <= 17`
- Use `in` to check membership: `"a" in "banana"` → `True`
- No need for `== True` with booleans: `if has_licence:` is cleaner than `if has_licence == True:`

## ✏️ Questions & Answers

### Q1 — Positive, Negative, or Zero
**Task:** Print `"positive"`, `"negative"`, or `"zero"` depending on `x`.

**Key concept:** Basic `if/elif/else` structure and indentation.

````python
x = -5

if x < 0:
    print("negative")  # Output: negative
elif x > 0:
    print("positive")
else:
    print("zero")
````

### Q2 — Teenager Check
**Task:** Print `"teenager"` if age is between 13 and 17 (inclusive), otherwise `"not a teenager"`.

**Key concept:** Inclusive bounds with `>=` and `<=`. Python also supports chained comparisons.

````python
age = 15

# Standard approach:
if age >= 13 and age <= 17:
    print("teenager")  # Output: teenager
else:
    print("not a teenager")

# Pythonic chained comparison (same result):
if 13 <= age <= 17:
    print("teenager")  # Output: teenager
else:
    print("not a teenager")
````

### Q3 — Login System
**Task:** Print `"access granted"` if username is `"admin"` AND password is `"1234"`, otherwise `"access denied"`.

**Key concept:** `==` for comparison (not `=`), and `and` logical operator.

````python
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("access granted")  # Output: access granted
else:
    print("access denied")
````

### Q4 — Score Validator
**Task:** Print `"invalid score"` if below 0 or above 100, `"pass"` if 50+, otherwise `"fail"`.

**Key concept:** Always check edge cases FIRST — Python stops at the first `True` condition.

````python
score = 85

if score < 0 or score > 100:  # ✅ Edge case checked FIRST
    print("invalid score")
elif score >= 50:
    print("pass")             # Output: pass
else:
    print("fail")

# ⚠️ Common mistake — checking pass/fail BEFORE invalid:
# if score >= 50:             → score of 110 would print "pass" incorrectly!
#     print("pass")
````

### Q5 — FizzBuzz
**Task:** Print `"fizzbuzz"` if divisible by both 3 and 5, `"fizz"` if by 3, `"buzz"` if by 5, else print the number.

**Key concept:** `%` (modulo) checks divisibility. Check the combined case first to avoid it being swallowed by `fizz` or `buzz`.

````python
n = 15

if n % 15 == 0:        # Divisible by both — checked FIRST
    print("fizzbuzz")  # Output: fizzbuzz
elif n % 3 == 0:
    print("fizz")
elif n % 5 == 0:
    print("buzz")
else:
    print(n)

# Alternative for fizzbuzz condition:
# if n % 3 == 0 and n % 5 == 0:
````

### Q6 — Letter in Word
**Task:** Print `"yes"` if the word contains the letter `"a"`, otherwise `"no"`.

**Key concept:** The `in` operator checks membership in strings, lists, and more.

````python
word = "banana"

if "a" in word:
    print("yes")  # Output: yes
else:
    print("no")

# 'in' also works with lists:
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)  # True
````

### Q7 — Both Even, Both Odd, or Mixed
**Task:** Print `"both even"`, `"both odd"`, or `"mixed"` based on two numbers.

**Key concept:** Combining `%` (modulo) with `and` to check multiple conditions.

````python
a = 4
b = 7

if a % 2 == 0 and b % 2 == 0:
    print("both even")
elif a % 2 != 0 and b % 2 != 0:
    print("both odd")
else:
    print("mixed")  # Output: mixed
````

### Q8 — Not Negative
**Task:** Print `"yes"` if the number is not negative, otherwise `"no"`.

**Key concept:** Two valid approaches — `>=` or using the `not` operator.

````python
x = 7

# Approach 1 — using >=:
if x >= 0:
    print("yes")  # Output: yes
else:
    print("no")

# Approach 2 — using 'not':
if not x < 0:
    print("yes")  # Output: yes
else:
    print("no")
````

### Q9 — Driving Eligibility
**Task:** Print `"eligible"` if age is 18+ AND has a valid licence, otherwise `"not eligible"`.

**Key concept:** No need for `== True` with booleans — use them directly.

````python
age = 20
has_licence = True

# ✅ Pythonic — use boolean directly:
if age >= 18 and has_licence:
    print("eligible")  # Output: eligible
else:
    print("not eligible")

# ⚠️ Works but unnecessary — avoid == True:
# if age >= 18 and has_licence == True:
````

### Q10 — Grade Calculator
**Task:** Print `"grade A"` for 90+, `"grade B"` for 80–89, `"grade C"` for 70–79, `"fail"` for below 70.

**Key concept:** Chained comparisons for range checks. Works naturally top-to-bottom.

````python
score = 85

if score >= 90:
    print("grade A")
elif 80 <= score <= 89:  # Chained comparison
    print("grade B")     # Output: grade B
elif 70 <= score <= 79:
    print("grade C")
else:
    print("fail")
````
